```python
import pandas as pd

# Load datasets
movies_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv')
credits_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv')

credits_df
```

```python
def extract_genres(genre_str):
    return ','.join(genre_str.split(','))  # Adjust this as needed based on actual data format

movies_df['genres'] = movies_df['genres'].apply(extract_genres)
```

```python
import ast

def extract_names(json_str):
    try:
        names = ast.literal_eval(json_str)
        return ','.join([item['name'] for item in names])
    except (SyntaxError, ValueError):
        return ''  # Return empty string or handle error as needed

credits_df['cast'] = credits_df['cast'].apply(lambda x: extract_names(x))
credits_df['crew'] = credits_df['crew'].apply(lambda x: extract_names(x) if x else '')
```

```python
# Merge movies_df with credits_df on the 'id' column
merged_df = movies_df.merge(credits_df,  left_on='id', right_on='movie_id')

merged_df
```

```python
def combine_tags(row):
    overview = row['overview'] if pd.notna(row['overview']) else ''
    genres = row['genres'] if pd.notna(row['genres']) else ''
    keywords = row['keywords'] if pd.notna(row['keywords']) else ''
    cast = row['cast'] if pd.notna(row['cast']) else ''
    crew = row['crew'] if pd.notna(row['crew']) else ''

    return ' '.join(overview.split() + 
                    genres.split(',') + 
                    keywords.split(',') + 
                    cast.split(',') + 
                    crew.split(','))

# Apply the function to create the tags column
merged_df['tags'] = merged_df.apply(combine_tags, axis=1)

# Clean the tags by replacing spaces if needed
merged_df['tags'] = merged_df['tags'].str.replace(' ', '')  # Adjust if necessary
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Vectorization
cv = CountVectorizer()
vectors = cv.fit_transform(merged_df['tags'])  # Keep it sparse

# Calculate cosine similarity
similarity = cosine_similarity(vectors)

```

```python
def recommend(movie):
    try:
        # Get the index of the movie
        movie_idx = merged_df[merged_df['original_title'] == movie].index[0]
        
        # Get the distances of the movies
        distances = similarity[movie_idx]
        
        # Get the top 5 similar movies (excluding the first one, which is the movie itself)
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

        # Use iloc to get the titles from merged_df
        recommendations = [merged_df.iloc[i[0]]['original_title'] for i in movie_list]
        
        return recommendations
    except IndexError:
        return []  # Handle case where the movie is not found
```

```python
print(recommend("Avatar"))
```
